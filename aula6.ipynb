```python
#Criação tupla
frutas = ('banana', 'uva', 'laranja')
print(frutas)
print(type(frutas))
```

```python
#Verificiar o tamnhao tupla
len(frutas)
```

```python
#Pegar o primeiro e o ultimo item
print(frutas[0])
print(frutas[-1])
```

```python
#Inserir novo item
frutas.append('tomate')
#Deletar um item da tupla
del frutas[0]
```

```python
tupla_del = ("ALALALALALAL", 1,3,4,'asdsdadsa')
print(tupla_del)

del tupla_del
print(tupla_del)
```

```python
minha_tupla = (1, 2, 3)
lista_resultante = list(minha_tupla)
print(lista_resultante)
print(type(lista_resultante))
```

```python
minha_lista = [4, 5, 6]
tupla_resultante = tuple(minha_lista)
print(tupla_resultante)
```

```python
tupla_novo = (1, 2, 3, 'Loteria', 'bonde do 7', True)
print(tupla_novo)
```

```python
tupla_frutas = ('maça','abacaxi','caju')
adicionar_tomate = 'tomate'
nova_tupla = tupla_frutas + (adicionar_tomate,)
print(nova_tupla)
```
