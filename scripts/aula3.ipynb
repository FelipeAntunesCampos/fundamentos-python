```python
#Declarar uma string
dia_da_semana = 'segunda'
mes = "maio"

print(dia_da_semana)
print(mes)
```

```python
#Usando a função len
print(len(dia_da_semana))
print(len(mes))
```

```python
#Primeira letra da string
print(dia_da_semana[0])

#ultima letra da string
print(dia_da_semana[-1])

#Pedaço do texto
print(dia_da_semana[0:4])

#juntar texto
print(dia_da_semana + ' é tristeza!!!')
```

```python
dia_da_semana = 'Segunda A dosodis 000000'

# Deixar em minuscula
print(dia_da_semana.lower())

# Deixar em maiúscula
print(dia_da_semana.upper())

# Deixar só a primeira letra maiúscula
print(dia_da_semana.capitalize())

# Lembrando que esses metodos nao alteraram a variavel
print(dia_da_semana)
```

```python
#quebrar texto
frase = 'A segunda acabou!'
print(frase)
print(type(frase))
frase_split = frase.split(' ')
print(frase_split)
print(type(frase_split))
```

```python
#quebrar texto
frase = '     A segunda acabou!     '
print(frase)
print(type(frase))
frase_split = frase.split(' ')
print(frase_split)
print(type(frase_split))
print('---------------------------')

#remover espaco no incio e no final
frase_strip = frase.strip()
print(frase_strip)
frase_split = frase_strip.split(' ')
print(frase_split)
print(type(frase_split))

```

```python
#concatenação de strings
nome = "Felipe"
idade = 17
print("Olá, " + nome + "! Você tem " + str(idade) + " anos.")

#Formatação de strings com %(Metodo antigo)
print("Olá, %s! Você tem %d anos." %(nome, idade ))
```

```python
#Método Format();
nome = "Felipe"
idade = 17
print("Olá, {}! Você tem {} anos.".format(nome, idade ))
```

```python
#Método F-strings (format Strings)
nome = "Felipe"
idade = 17
print(f"Olá, {nome}! Você tem {idade} anos.")
```
