```python
#Declarar uma lista
frutas = ['banana', 'uva', 'manga']
print(frutas)
print(type(frutas))
```

```python
#Pegar os itens da lista por posicao
print(frutas[0])
print('----------')
print(frutas[1])
```

```python
#declarar outros tipos na lista
a_lista = ['banana',None,'uva','goiaba', 4==4]

#4==4 verificar se 4 é igual a 4

print(a_lista)
```

```python
#colocar uma lista em uma lista
a_list = [23,None, 3.14, frutas, 3 <= 5]

print(a_list)
print(a_list[3][2])
```

```python
#Criando uma matriz 3x3
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]
print(matriz)
print(type(matriz))
print(matriz[1][2])
```

```python
#Modificando elementos da matriz (provando que lista é mutável)
matriz[0][0] = 10
matriz[1][2] = 'a'

print(matriz)
```

```python
#declarar lista vazia
lista_vazia = []
print(lista_vazia)
```

```python
#Utilizar o len
print(len(lista_vazia))
print('--------')
print(len(frutas))
```

```python
#pegar pedaço da lista
print(a_lista)
print('\n')
print(a_lista[:2])
print('\n')
print(a_lista[2:4])
```

```python
#Inserção de dados na lista
frutas.insert(1, 'melão')

print(frutas)
```

```python
#Inserção de dados no final da lista
frutas.append('maça')

print(frutas)
```

```python
#lista de frutas adicionais
frutas_adicionas = ['caju', 'morango', 'abacaxi']

#estendendo a lista de frutas com a lista de frutas adicionais
frutas.extend(frutas_adicionas)

print(frutas)
```

```python
frutas.append(['abacaxi', 'uva'])
print(frutas)
```

```python
#Remover um elemento da lista
numeros = [1, 2, 3, 4, 3, 5]

# Removendo o número 3 da lista
numeros.remove(3)

print(numeros)
print(frutas)
```

```python
#existe o pop para remover e adicionar a uma variavel
print(frutas)
fruta_removida = frutas.pop()
print(frutas)
print(fruta_removida)
```

```python
#juntar listas
mais_frutas = frutas + ['morango'] + ['Goiaba', 'tangerina']

print(mais_frutas)
```

```python
#Copia do dados
mais_frutas_copia = mais_frutas.copy()

print(mais_frutas_copia)

#modificar lista
mais_frutas_copia.remove('Goiaba') # remover uma item
mais_frutas_copia.pop() #remover o último item da lista
print(mais_frutas_copia)
print(mais_frutas)
```
