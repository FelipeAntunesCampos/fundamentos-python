```python
#Declarando o dicionario
person1 = {
    'nome': 'Felipe',
    'sexo': 'Masculino',
    'idade': 30,
    'casado': False
}
print(person1)
print(type(person1))
```

```python
#Usar a funcao dict
person2 = dict(nome='Maria', sexo='Feminino', idade=29, casado=True)
print(person2)
print(type(person2))
print(person2['sexo']) #Buscando pela chave do dicionario
print(person2['idade'])
```

```python
dict1 = {}
print(dict1)
dict1['chave1'] = 'não esta mais vazio!'
print(dict1)
```

```python
dict2 = {'Pessoa1':['Felipe',30,'Aluno'],
         'Pessoa2':['julia',50,'Medica']}
print(dict2)
print(dict2['Pessoa1'][0])
```

```python
#dicionario alinhado
pessoa = {
    'nome': 'Jota',
    'idade': 30,
    'endereco': {
      'rua': 'Rua Principal',
      'cidade': 'Cidade A',
      'estado': 'Estado X'
    }
}
print(pessoa['nome'], pessoa['idade'], pessoa['endereco']['rua'], pessoa['endereco']['cidade'], pessoa['endereco']['estado'])
```

```python
#Você também pode usar o método get para acessar o valor
pessoa.get('nome')
```

```python
#alterar a informação já criada
pessoa['casado'] = True
#adicionar um novo campo e chave
pessoa['Nacionalidade'] = 'Brasileiro'

print(pessoa)
print(type(pessoa))
```

```python
#Para remover uma chave e o valor associado de um dicionário, use o método
print(pessoa)
pessoa.pop('Nacionalidade')
print(pessoa)
```

```python
#Os dicionários também fornece métodos para visualizar as chaves
print(pessoa.keys())
```

```python
#Pegar todos os valores ,ou seja, pegar os valores dentro das chaves
print(pessoa.values())
```

```python
#Retorna uma view de todos os itens = dicionar.view
#Criando um dicionário
meu_dicionario = {'a': 1, 'b': 2, 'c': 3}

#Exibindo o dicionário original
print('Dicionário original:', meu_dicionario)

#Usando o método clear para remover todos itens
meu_dicionario.clear()

#Exibindo o dicionário após o clear
print('Dicionário após clear:', meu_dicionario)
```

```python
#Criando um dicionário
meu_dicionario = {'a': 1, 'b': 2, 'c': 3}

#Exibindo o dicionário original
print('Dicionário original:', meu_dicionario)

#Usando o método del para remover o item associado à chave 'b'
del meu_dicionario['b']

#Exibindo o dicionário após o clear
print('Dicionário após deletar "b":', meu_dicionario)
```

```python
#Dicionario original
dicionario1 = {'a': 1, 'b': 2}

#ADcionando elementos de outro dicionário
dicionario2 = {'b':3, 'c': 4}
dicionario1.update(dicionario2)

#exibindo o dicionário após a atualização
print('Dicionário após update:', dicionario1)
```

```python
informacoes_pessoa = {'nome':'Maria',
                      'idade': 28,
                      'cidade':'São paulo'}
print(informacoes_pessoa)
print(informacoes_pessoa['nome'],informacoes_pessoa['idade'], informacoes_pessoa['cidade'])
```

```python
estoque = {'Maçãs':10,'Bananas':15,'Laranjas':8}
estoque2 = {'Maçãs':20, 'Bananas':10,'Laranjas':9}
print('O Primeiro estoque contém' + str(estoque))
estoque.update(estoque2)
print('O estoque atualizado contém',str(estoque))

```
