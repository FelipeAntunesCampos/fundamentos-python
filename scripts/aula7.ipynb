```python
#Switch case
nacionalidade = input("Digite sua nacionalidade (brasileiro, americano, canadense): ")

match nacionalidade:
  case "brasileiro":
      print("Você é do Brasil")
  case "americano":
      print("Você é dos Estados Unidos")
  case "canadense":
      print("Você é do Canadá")
  case _:
      print("Nacionalidade nao reconhecida")
```
