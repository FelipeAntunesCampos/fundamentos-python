```python
#Pode retornar um valor
def soma(a,b,c):
  return a+b+c

print(soma(4,6,4))
```

```python
#pode retornar mais de um valor
def soma_multiplicacao(a,b,c):
  return (a+b+c), (a*b*c)

soma, multiplicacao = soma_multiplicacao(1,1,1)
print("A soma é {} e a multiplicacao é {}".format(soma,multiplicacao))
```

```python
def soma(x,y):
  return x + y

print(soma(3,4))
```

```python
soma = lambda x, y: x + y
print(soma(3, 4))
```

```python
numeros = [1, 2, 3, 4, 5]
ao_quadrado = list(map(lambda x: x**2, numeros))
print(ao_quadrado)
```

```python
numeros = [1, 2, 3, 4, 5]
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)
```
